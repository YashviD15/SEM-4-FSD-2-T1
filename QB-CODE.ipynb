```python
#209
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
f=pd.read_csv("car data.csv")
z=pd.DataFrame(f)
z["Age_car"]=2023-z["Year"]
x=z[["Age_car","Kms_Driven","Fuel_Type","Seller_Type","Transmission"]]
y=z["Selling_Price"]
x=pd.get_dummies(x,drop_first=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
pre=model.predict(x_test)
rs=r2_score(y_test,pre)
mse=mean_squared_error(y_test,pre)
print("r2square:",rs)
print("mean_squared_error:",mse)
```

```python
import matplotlib.pyplot as plt
f=pd.read_csv("car data.csv")
z=pd.DataFrame(f)
z["Age_car"]=2023-z["Year"]
x=z[["Age_car","Kms_Driven","Fuel_Type","Seller_Type","Transmission"]]
y=z["Selling_Price"]
x=pd.get_dummies(x,drop_first=True)
xy=range(0,42)
r2=[]
for i in xy:
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    model=LinearRegression()
    model.fit(x_train,y_train)
    pre=model.predict(x_test)
    rs=r2_score(y_test,pre)
    r2.append(rs)
r2
plt.plot(xy,r2)
plt.show()
```

```python
#252
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.neighbors import KNeighborsClassifier
f=pd.read_csv("drug200.csv")
z=pd.DataFrame(f)
x=z[["Age","Sex","BP","Cholesterol"]]
y=z["Drug"]
x=pd.get_dummies(x,drop_first=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
pre=model.predict(x_test)
cm=confusion_matrix(y_test,pre)
acc=accuracy_score(y_test,pre)
cr=classification_report(y_test,pre)
print(pre)
print("classification_report",cr)
print("accuracy_score",acc)
print("confusion_matrix",cm)
print(model.classes_)
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(data=cm,annot=True)

```
